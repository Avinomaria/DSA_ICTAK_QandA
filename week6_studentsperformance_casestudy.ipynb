# 2. What do you think about the students' parental level of education?1

```python
parent_edu=df["parental level of education"].value_counts()
parent_edu
```

```python
ax=sns.countplot(x='parental level of education', data=df)
plt.grid(linestyle='dotted')
plt.title('Parental Level of Education',fontsize=10,fontweight='bold')
plt.xlabel('Level of Education',fontsize=10,fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel('Count',fontweight='bold')
plt.show()
```

Can conclude that most of the parents are either in some college or associate's degree. Only a few numbers are with master's degree.
