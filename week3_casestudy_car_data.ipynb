# 3. Bar Graph of Male vs Female Buyers

```python
ax =df['Buyer Gender'].value_counts()
ax
```

> obtained the number of female and male buyers

```python
c=['orange','blue']
ax = df['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0,color=c)
plt.xlabel("Gender")
plt.ylabel("Number of Buyers")
plt.title("Bar Graph of Gender" )
plt.grid(linestyle='dotted')
ax.set_xticklabels(('Female', 'Male'))
```

# 4.Top 5 Cars Based on Sale Price
